```python
def checkElementHelper(x, y):
    for item in y:
        if item[0] == x:
            return True
    return False
```

```python
def union(setA, setB):
    result = []
    for elem in setA:
        result.append(elem)
    
    for elem in setB:
        if not checkElementHelper(elem[0], result):
            result.append(elem)
        else:
            for i in range(len(result)):
                if result[i][0] == elem[0]:
                    result[i] = (elem[0], max(result[i][1], elem[1]))
    return result
```

```python
def intersection(setA, setB):
    result = []
    for elem in setA:
        if checkElementHelper(elem[0], setB):
            for item in setB:
                if item[0] == elem[0]:
                    result.append((elem[0], min(elem[1], item[1])))
    return result
```

```python
def complement(setA):
    result = []
    for elem in setA:
        result.append((elem[0], 1 - elem[1]))
    return result
```

```python
def fuzzy_sum(setA, setB):
    result = []
    for elem in setA:
        found = False
        for item in setB:
            if item[0] == elem[0]:
                a, b = elem[1], item[1]
                result.append((elem[0], a + b - (a * b)))
                found = True
        if not found:
            result.append(elem)
    return result
```

```python
def fuzzy_product(setA, setB):
    result = []
    for elem in setA:
        for item in setB:
            if item[0] == elem[0]:
                result.append((elem[0], elem[1] * item[1]))
    return result
```

```python
def bounded_sum(setA, setB):
    result = []
    for elem in setA:
        found = False
        for item in setB:
            if item[0] == elem[0]:
                result.append((elem[0], min(1, elem[1] + item[1])))
                found = True
        if not found:
            result.append(elem)
    return result
```

```python
def bounded_diff(setA, setB):
    result = []
    for elem in setA:
        found = False
        for item in setB:
            if item[0] == elem[0]:
                result.append((elem[0], max(0, elem[1] - item[1])))
                found = True
        if not found:
            result.append(elem)
    return result
```

```python
A = [('x1', 0.8), ('x2', 0.6), ('x3', 0.4)]
B = [('x1', 0.5), ('x2', 0.9), ('x4', 0.7)]
```

```python
union(A, B)
```

```python
intersection(A,B)
```

```python
complement(A)
```

```python
fuzzy_sum(A,B)
```

```python
fuzzy_product(A,B)
```

```python
bounded_sum(A,B)
```

```python
bounded_diff(A,B)
```

```python
print(" Commutativity  Equal?", sorted(union(A, B)) == sorted(union(B, A)))
```

```python
print(" Involution  Equal?", sorted(complement(complement(A))) == sorted(A))
```

```python
print(" Idempotency  Equal?", sorted(union(A, A)) == sorted(A))
```
